```python
import os
from PIL import Image
import shutil

def convert_images_to_jpg(folder_path, output_folder, quality=95):
    """
    Copy all photos and convert only PNG files to JPG format.
    JPG files are copied as-is, other formats are also copied.
    
    Args:
        folder_path (str): Path to the folder containing images
        output_folder (str): Path to output folder
        quality (int): JPG quality (1-100, default 95)
    """
    
    # Supported image formats
    image_formats = {'.png', '.jpg', '.jpeg', '.bmp', '.tiff', '.tif', '.webp', '.gif', '.ico', '.ppm', '.pgm', '.pbm'}
    
    # Check if input folder exists
    if not os.path.exists(folder_path):
        print(f"Error: Input folder '{folder_path}' does not exist.")
        return
    
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    converted_count = 0
    copied_count = 0
    error_count = 0
    
    print(f"Processing images in: {folder_path}")
    print(f"Output folder: {output_folder}")
    print("-" * 50)
    
    # Process all files in the folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        
        # Skip if not a file
        if not os.path.isfile(file_path):
            continue
        
        # Get file extension
        file_ext = os.path.splitext(filename)[1].lower()
        
        # Skip if not an image format
        if file_ext not in image_formats:
            continue
        
        try:
            if file_ext == '.png':
                # Convert PNG to JPG
                with Image.open(file_path) as img:
                    # Convert to RGB (handle transparency)
                    if img.mode in ('RGBA', 'LA'):
                        # Create white background for transparent images
                        rgb_img = Image.new('RGB', img.size, (255, 255, 255))
                        rgb_img.paste(img, mask=img.split()[-1])
                        img = rgb_img
                    elif img.mode == 'P':
                        img = img.convert('RGBA')
                        rgb_img = Image.new('RGB', img.size, (255, 255, 255))
                        rgb_img.paste(img, mask=img.split()[-1] if len(img.split()) > 3 else None)
                        img = rgb_img
                    elif img.mode != 'RGB':
                        img = img.convert('RGB')
                    
                    # Create output filename (change extension to .jpg)
                    base_name = os.path.splitext(filename)[0]
                    output_filename = f"{base_name}.jpg"
                    output_path = os.path.join(output_folder, output_filename)
                    
                    # Save as JPG
                    img.save(output_path, 'JPEG', quality=quality, optimize=True)
                    
                    print(f"✓ Converted: {filename} → {output_filename}")
                    converted_count += 1
                    
            else:
                # Copy all other image files as-is (including JPG files)
                output_path = os.path.join(output_folder, filename)
                shutil.copy2(file_path, output_path)
                
                print(f"✓ Copied: {filename}")
                copied_count += 1
                
        except Exception as e:
            print(f"✗ Error processing {filename}: {str(e)}")
            error_count += 1
    
    print("-" * 50)
    print(f"Processing complete!")
    print(f"PNG files converted to JPG: {converted_count}")
    print(f"Other image files copied: {copied_count}")
    if error_count > 0:
        print(f"Errors encountered: {error_count} files")

def main():
    """Main function to run the converter with user input."""
    
    print("Image to JPG Converter")
    print("=" * 30)
    
    # Get input folder path from user
    folder_path = input("Enter the input folder path containing images: ").strip()
    
    # Get output folder path from user
    output_folder = input("Enter the output folder path: ").strip()
    
    # Optional: Ask for quality
    quality_input = input("Enter JPG quality 1-100 (press Enter for 95): ").strip()
    quality = int(quality_input) if quality_input.isdigit() and 1 <= int(quality_input) <= 100 else 95
    
    convert_images_to_jpg(folder_path, output_folder, quality)

if __name__ == "__main__":
    main()
```
